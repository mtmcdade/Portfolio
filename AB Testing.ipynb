```python
#Load packages
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import ttest_ind

#load datasets
control = pd.read_csv("control_group.csv", sep=';')
test = pd.read_csv("test_group.csv", sep=';')

#check the first 5 rows
print("Control")
print(control.head())
print("Test")
print(test.head())
```

```python
#Cleaning data
#Check datatypes
print("Control data types:")
print(control.dtypes)
print("\nTest data types:")
print(test.dtypes)

#Fill nulls with 0
control = control.fillna(0)
test = test.fillna(0)
```

```python
#Testing impact on $ spent
print("Control mean purchases:", control['Spend [USD]'].mean())
print("Test mean purchases:", test['Spend [USD]'].mean())

#Comparing Means
control_mean= control['Spend [USD]'].mean()
test_mean= test['Spend [USD]'].mean()
print("Control mean $ Spent:",control_mean)
print("Test mean $ Spent:", test_mean)

#Compare graphically
labels = ['Control', 'Test']
means = [control_mean, test_mean]
plt.figure(figsize=(6, 4))
plt.bar(labels, means, color=['#4F81BD', '#E69F00'])
plt.ylabel('Mean $ Spent')
plt.title('Mean $ Spent: Control vs Test Group')
plt.show()

# Perform t-test on the arrays
t_stat, p_value = ttest_ind(control['Spend [USD]'], test['Spend [USD]'])
print("t-statistic:", t_stat)
print("p-value:", p_value)

#Showing significance based on p value >= 0.5
if pd.isna(p_value):
    print("Result: The test could not be performed (p-value is NaN). Check your data for missing or constant values.")
elif p_value < 0.05:
    print("Result: The difference in Spend [USD] is statistically significant (p < 0.05).")
else:
    print("Result: The difference in Spend [USD] is not statistically significant (p >= 0.05).")
```
